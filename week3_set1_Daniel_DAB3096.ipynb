# **OUTLIERS**

**OUTLIERS IN AGE COLUMN**

```python
# Age column after  cleaning
set_one['Age']
```

```python
# calculating the outliers
# calulating Q1 (25th quartile) and Q3 (75th quartile)
Q1 = set_one['Age'].quantile(0.25)
Q3 = set_one['Age'].quantile(0.75)
print(Q1)
print(Q3)
```

```python
# calculating interquartile
IQR = Q3-Q1
print(IQR)
```

```python
# Now calculating upper and lower bound
lower_bound = Q1-1.5*IQR
upper_bound = Q3+1.5*IQR
print(lower_bound)
print(upper_bound)
```

```python
# calculating actual outliers
the_outliers = set_one['Age'][(set_one['Age']<lower_bound) | (set_one['Age']>upper_bound)]
print(the_outliers)
```

**OUTLIERS IN SALARY COLUMN**

```python
# calculating the outliers in minimum salary columnn
# calulating Q1 (25th quartile) and Q3 (75th quartile)
Q1_salary_min = set_one['Minimum_Salary'].quantile(0.25)
Q3_salary_min = set_one['Minimum_Salary'].quantile(0.75)
print(Q1_salary_min)
print(Q3_salary_min)
print('IQR:', Q3_salary_min-Q1_salary_min)
```

```python
# calculating the outliers in maximum salary columnn
# calulating Q1 (25th quartile) and Q3 (75th quartile)
Q1_salary_max = set_one['Maximum_Salary'].quantile(0.25)
Q3_salary_max = set_one['Maximum_Salary'].quantile(0.75)
print(Q1_salary_max)
print(Q3_salary_max)
IQR_max_sal = Q3_salary_max-Q1_salary_max
print(IQR_max_sal)
```

```python
# Now calculating upper and lower bound
lower_bound_max_sal = Q1_salary_max-1.5*IQR_max_sal
upper_bound_max_sal = Q3_salary_max+1.5*IQR_max_sal
print(lower_bound_max_sal)
print(upper_bound_max_sal)
```

```python
# calculating actual outliers
the_outliers_max_sal = set_one['Maximum_Salary'][(set_one['Maximum_Salary']<lower_bound_max_sal) | (set_one['Age']>upper_bound_max_sal)]
print(the_outliers_max_sal)
```
